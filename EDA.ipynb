```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Lista de nombres de DataFrames para referencia
nombres = ['acc_vel_loc_sinrangos', 'velocidad_sin_Rangos', 'velocidad_por_prov', 'totales_vmd', 
           'accesos_tecnologia_localidad', 'totales_accesos_por_tecnologia', 'accesos_por_tecnologia', 
           'dial_baf', 'totales_dial_baf', 'penetracion_poblacion', 'penetracion_hogares', 'penetracion_totales', 
           'totales_accesos_por_rango', 'accesos_por_rango', 'ingresos']

# Función para identificar y tratar outliers solo en columnas numéricas
def tratar_outliers(df):
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Reemplazar outliers con NaN
        df.loc[(df[column] < lower_bound) | (df[column] > upper_bound), column] = np.nan
    # Rellenar NaN con la mediana de cada columna numérica
    df.fillna(df.select_dtypes(include=[np.number]).median(), inplace=True)
    return df

# Aplicar la función a cada DataFrame
for i, df in enumerate(dataframes):
    dataframes[i] = tratar_outliers(df)

# Visualizar los outliers con box plots
for i, df in enumerate(dataframes):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df.select_dtypes(include=[np.number]))
    plt.title(f'Outliers en {nombres[i]}')
    plt.show()
```

```python
##Grafica todas las variables para su analisis y escoger unas mas especificas 
# Función para graficar variables cuantitativas
def graficar_cuantitativas(df, nombre):
    cols_cuantitativas = df.select_dtypes(include=[np.number]).columns
    for col in cols_cuantitativas:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[col], kde=True)
        plt.title(f'Distribución de {col} en {nombre}')
        plt.show()
        
        plt.figure(figsize=(10, 6))
        sns.boxplot(y=df[col])
        plt.title(f'Box Plot de {col} en {nombre}')
        plt.show()

# Función para graficar variables cualitativas
def graficar_cualitativas(df, nombre):
    cols_cualitativas = df.select_dtypes(include=['object']).columns
    for col in cols_cualitativas:
        plt.figure(figsize=(10, 6))
        sns.countplot(y=df[col], order=df[col].value_counts().index)
        plt.title(f'Frecuencia de {col} en {nombre}')
        plt.show()

# Aplicar las funciones a cada DataFrame
for nombre, df in zip(nombres, dataframes):
    graficar_cuantitativas(df, nombre)
    graficar_cualitativas(df, nombre)

```

```python

# Función para graficar variables cuantitativas seleccionadas
def graficar_cuantitativas_seleccionadas(df, nombre, columnas):
    for col in columnas:
        if col in df.columns:
            plt.figure(figsize=(10, 6))
            sns.histplot(df[col], kde=True)
            plt.title(f'Distribución de {col} en {nombre}')
            plt.show()
            
            plt.figure(figsize=(10, 6))
            sns.boxplot(y=df[col])
            plt.title(f'Box Plot de {col} en {nombre}')
            plt.show()

# Función para graficar variables cualitativas seleccionadas
def graficar_cualitativas_seleccionadas(df, nombre, columnas):
    for col in columnas:
        if col in df.columns:
            plt.figure(figsize=(10, 6))
            sns.countplot(y=df[col], order=df[col].value_counts().index)
            plt.title(f'Frecuencia de {col} en {nombre}')
            plt.show()

# Listas de columnas relevantes para cada DataFrame
columnas_cuantitativas = {
    'acc_vel_loc_sinrangos': ['Velocidad (Mbps)', 'Accesos'],
    'velocidad_sin_Rangos': ['Velocidad', 'Accesos'],
    'accesos_tecnologia_localidad': ['Accesos'],
    'velocidad_por_prov': ['Mbps (Media de bajada)'],
    'totales_vmd': ['Total'],
    'penetracion_poblacion': ['Accesos por cada 100 hab'],
    'penetracion_hogares': ['Accesos por cada 100 hogares'],
    'ingresos': ['Ingresos (miles de pesos)'],
    'totales_accesos_por_rango': ['Hasta 512 kbps', 'Entre 512 Kbps y 1 Mbps', 'Entre 1 Mbps y 6 Mbps', 'Entre 6 Mbps y 10 Mbps', 'Más de 30 Mbps']
}

columnas_cualitativas = {
    'acc_vel_loc_sinrangos': ['Partido'],
    'velocidad_sin_Rangos': ['Provincia'],
    'accesos_tecnologia_localidad': ['Tecnología'],
    'penetracion_hogares': ['Trimestre'],
    'ingresos': ['Trimestre']
}

# Diccionarios de DataFrames para referencia
nombres = ['acc_vel_loc_sinrangos', 'velocidad_sin_Rangos', 'accesos_tecnologia_localidad',
           'velocidad_por_prov', 'totales_vmd', 'penetracion_poblacion', 'penetracion_hogares', 
           'ingresos', 'totales_accesos_por_rango']

dataframes = [acc_vel_loc_sinrangos, velocidad_sin_Rangos, accesos_tecnologia_localidad, 
              velocidad_por_prov, totales_vmd, penetracion_poblacion, penetracion_hogares, 
              ingresos, totales_accesos_por_rango]

# Graficar las variables cuantitativas seleccionadas
for nombre, df in zip(nombres, dataframes):
    if nombre in columnas_cuantitativas:
        graficar_cuantitativas_seleccionadas(df, nombre, columnas_cuantitativas[nombre])

```
